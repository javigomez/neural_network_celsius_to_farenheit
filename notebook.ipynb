```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0 , 8, 15, 22, 38], dtype=float)
farenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("starting training...")
historial= modelo.fit(celsius,farenheit,epochs=1000,verbose=False)
print("model is trained!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoch")
plt.ylabel("Loss magnitude")
plt.plot(historial.history["loss"])
```

```python
grados_celsius_a_convertir = 100.0
prediccion = modelo.predict([grados_celsius_a_convertir])
print(str(grados_celsius_a_convertir) + "C son" + str(prediccion) + "grados Farenheit")
```

```python
print("let's see the internal weights that the artificial network used")
print(capa.get_weights())
```

the Celsius to Farenheit formula is: 
`farenheit = celsius * 1.8 + 32`

The network calculated on itself:
`farenheit = celsius * 1.7981305 + 31.931042`

It nearly got it!
